# II. Цілі числа та числа з рухомою комою

3. Створити дві змінні, що посилаються на будь-які ціле число та число з рухомою комою та продемонструвати такі арифметичні операції: додавання, віднімання, ділення, множення, ділення без залишку, ділення по модулю, приведення до ступеню. Всі результати операцій вивести на екран.

```python
num_int = 10
num_float = 3.5

print(f"Додавання: {num_int + num_float}")
print(f"Віднімання: {num_int - num_float}")
print(f"Ділення: {num_int / num_float}")
print(f"Множення: {num_int * num_float}")
print(f"Ділення без залишку: {num_int // num_float}")
print(f"Ділення по модулю: {num_int % num_float}")
print(f"Приведення до ступеню: {num_int ** num_float}")
```

```python
# Використовуємо змінні з попередньої вправи
converted_float = int(num_float)
print(f"Float {num_float} -> Int: {converted_float}")

# Булеве значення до цілого
my_bool = True
converted_bool = int(my_bool)
print(f"Bool {my_bool} -> Int: {converted_bool}")
```

18. Створити множину, маючи список my_list = [1, 1, 2, 67, 67, 8, 9]. Пояснити, чому "зникли" деякі елементи.

```python
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

print(f"Об'єднання: {set_a.union(set_b)}")
print(f"Різниця (a-b): {set_a.difference(set_b)}")
print(f"Перетин: {set_a.intersection(set_b)}")
print(f"Симетрична різниця: {set_a.symmetric_difference(set_b)}")
```
